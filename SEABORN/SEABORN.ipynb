```python
d = pd.read_csv("dünyam.csv")
d.dropna(how="any",inplace=True) #Nan olan değerleri atar.
d
d.head()
```

```python
sns.set(style="dark",palette="pastel",font_scale=1.5) #Arka kısmını değiştirir. palette seborn sitesinden baktık.grafiği değiştirir.
#font_scale kenardaki sayıların boyutunu belirler.

sns.distplot(d["Doğum oranı"],bins=20,kde=False,color='r') #bins=grafikteki kutucukları ayırır. kde=üzerine çizgi çizer.
plt.tight_layout() #Ottomatik daha düzgün bir görüntü verir.
```

```python
sns.set(style="darkgrid",palette="muted",font_scale=1.5)

sns.distplot(d["Yıllık değişim oranı"],hist=False,rug=True,color="y") #hist kolonları oluşturur. rıg alttaki çizgileri oluşturur.
plt.tight_layout()
```

```python
sns.set(style="white",palette="bright",font_scale=1.5)

sns.distplot(d["Ortalama yaş"],hist=False,color="g",kde_kws={"shade":True}) #kde_kws={"shade":True} çizginin içini doldurur.
plt.tight_layout()
```

```python
sns.set(style="whitegrid", palette="Blues",font_scale=1.5)
sns.distplot(d["Nüfüs"],color='m')
plt.tight_layout()
```

# PAIRGRID()

```python
m=sns.PairGrid(d) #Tablolayla ilgili ilişkili tablolar oluşturdu.
m.map_diag(sns.distplot)
m.map_upper(plt.scatter) #Yuakrıdakileri oluşturur.
m.map_lower(sns.kdeplot) #alttakileri oluşturur.
```
